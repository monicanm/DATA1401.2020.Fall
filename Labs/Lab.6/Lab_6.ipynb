    * `eye(n)`: returns the n by n identity matrix.

4. Add the following member functions to your class. Make sure to appropriately test the dimensions of the matrices to make sure the operations are correct.
    * `M.scalarmul(c)`: a matrix that is scalar product $cM$, where every element of $M$ is multiplied by $c$.
    * `M.add(N)`: adds two matrices $M$ and $N$. Donâ€™t forget to test that the sizes of the matrices are compatible for this and all other operations.
    * `M.sub(N)`: subtracts two matrices $M$ and $N$.
    * `M.mat_mult(N)`: returns a matrix that is the matrix product of two matrices $M$ and $N$.
    * `M.element_mult(N)`: returns a matrix that is the element-wise product of two matrices $M$ and $N$.
    * `M.equals(N)`: returns true/false if $M==N$.

```python
class Matrix():
    def constant(n,m,c):
      out = []
      for i in range(1, (m*c)+1):
          matrix_list = []
          for j in range(1, n+1):
              x = (i*j)*(c*1.0)
              matrix_list.append(x)
          out.append(matrix_list)
      return out
    def zeros(m,n):
        out=[]
        for i in range(m):
            row=[]
            for j in range(n):
                row.append(0.)
            out.append(row)
        return out
    def ones(m,n):
        out=[]
        for i in range(m):
            row=[]
            for j in range(n):
                row.append(1.)
            out.append(row)
        return out    
    def eye(n):
        out=[]
        for i in range(n):
            row=[]
            for j in range(n):
                eye=n
                row.append(eye)
            out.append(row)
        return out

    def scalarmul(M,c):
        m,n=matrix.shape(M)
        M_out=Matrix.zeros(m,n)
        for i in range(m):
            for j in range(n):
                M_out[i][j]=c*M[i][j]
        return M_out

    def add(M,N):
        m1,n1=matrix.shape(M)
        m2,n2=matrix.shape(N)
        if m1==m2 and n1==n2:
            MN=Matrix.zeros(m1,n1)
            for i in range(m1):
                for j in range(n1):
                    MN[i][j]=M[i][j]+N[i][j]
            return MN
        return False

    def sub(M,N):
        m1,n1=matrix.shape(M)
        m2,n2=matrix.shape(N)
        if m1==m2 and n1==n2:
            MN=Matrix.zeros(m1,n1)
            for i in range(m1):
                for j in range(n1):
                    MN[i][j]=((M[i][j])-(N[i][j]))
            return MN
        return False

    def mat_mult(M,N):
        m1,n1=matrix.shape(M)
        m2,n2=matrix.shape(N)
        if m1==m2 and n1==n2:
            M3=Matrix.zeros(m1,n1)
            for i in range(m1):
                for j in range(n1):
                    M3[i][j]=M[i][j]*N[i][j]
            return M3
        return False

    def neg(M):
        return Matrix.mat_mult(M,-1)
    
    def element_mult(M,N):
        m1,n1=matrix.shape(M)
        m2,n2=matrix.shape(N)
        if m1==m2:
            M3=Matrix.zeros(m1,n2)
            for i in range(m1):
                for j in range(n2):
                    for k in range(m1):
                        M3[i][j]+=M[i][k]*N[k][j]
            return M3
        return False

    def equals(M):
        if isinstance(M,list):
            row_length=len(M[0])
            for row in M:
                if not row_length==len(row):
                    return False
        else:
            False
        return True
```
